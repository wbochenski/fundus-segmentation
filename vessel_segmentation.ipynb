```python
from pathlib import Path

from PIL import Image

from src.models.gabriel_lepetitaimon_fundus_vessel import predict
from src.utils import imshow
```

```python
image_paths = sorted(Path("data/raw/images").glob("*"))
target_paths = sorted(Path("data/raw/target").glob("*"))
pairs = list(zip(image_paths, target_paths))[:3]

for image_path, target_path in pairs:
    image = Image.open(image_path).convert("RGB")
    target = Image.open(target_path).convert("L")
    predicted = predict(image_path)

    imshow(image, target, predicted)
```
